```python
import main
import models.HAR_GK
import models.garch_model
import models.path_dependent_volatility_model
import models.RF_model
import models.BS_model
import importlib

## We can remove this in the final version, it helps me work on the main file and this simultaneously
importlib.reload(models.HAR_GK)
importlib.reload(models.garch_model)
importlib.reload(models.path_dependent_volatility_model)
importlib.reload(models.RF_model)
importlib.reload(models.BS_model)
importlib.reload(main)


tickers = ['VGT', 'AAPL', 'MSFT', 'NVDA']
start_date = '2019-01-01'
end_date = '2021-01-01'
start_date_predict = end_date
end_date_predict = '2021-03-01'

## The first variable is str for how you want the label to appear
## The second variable is the name of the model file. 

models_names = {"HAR-GK": models.HAR_GK   , "Path Dependent": models.path_dependent_volatility_model} #, "Black-Scholes": models.BS_model}

### all_results stores the MSE(mean squared error) and QLIKE(Quasi-likelihood) for each ticker and model combination.
all_results = {}

## Running the loop for each ticker

for ticker in tickers:

    ## Calling the download_data class tp get the data for the ticker and date range.

    df = main.dowloand_data(ticker, start_date, end_date, start_date_predict, end_date_predict)
    data = df.training_data()

    ## Running the for each model
    model_vol  = {}

    for model_name in models_names:
        fits = models_names[model_name].fit(data, start_date_predict, end_date_predict)
        predicted_volatility = fits.test()

        model_vol[model_name] = predicted_volatility
        print(f"Model: {model_name}, Ticker: {ticker}, Predicted Volatility: {predicted_volatility}")

    Compare = main.compare_models(ticker, model_vol, df.rolling_volatility(), start_date_predict, end_date_predict)

    Compare.plot_models()
    all_results[(ticker, "MSE")] = Compare.MSE()
    all_results[(ticker, "QLIKE")] = Compare.QLIKE()



main.plot_table_all(all_results)


```
