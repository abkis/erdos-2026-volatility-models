The Black-Scholes model assumes volatility is **constant**. We estimate that constant from the training window and hold it fixed across the entire prediction period. This gives the constant-volatility **baseline** that HAR-GK, GARCH, and other models are benchmarked against.

```python
import main
import BS
import importlib

## We can remove this in the final version, it helps me work on the main file and this simultaneously
importlib.reload(BS)
importlib.reload(main)


tickers = ['VGT', 'AAPL', 'MSFT', 'NVDA']
start_date = '2019-01-01'
end_date = '2021-01-01'
start_date_predict = '2021-01-01'
end_date_predict = '2021-03-01'

## The first variable is str for how you want the label to appear
## The second variable is the name of the model file.
models_names = [["BS", BS]]

### all_results stores the MSE(mean squared error) and QLIKE(Quasi-likelihood) for each ticker and model combination.
all_results = {}

## Running the loop for each ticker

for ticker in tickers:

    ## Calling the download_data class to get the data for the ticker and date range.

    df = main.dowloand_data(ticker, start_date, end_date, start_date_predict, end_date_predict)
    data = df.training_data()

    ## Running the loop for each model
    models_predicted = []
    for str_name, model_name in models_names:
        fits = model_name.fit(data, start_date_predict, end_date_predict)
        predicted_volatility = fits.test()
        models_predicted.append([predicted_volatility])

        models = [[str_name, predicted_volatility]]

    Compare = main.compare_models(ticker, models, df.rolling_volatility(), start_date_predict, end_date_predict)

    Compare.plot_models()
    Compare.plot_tables()
    all_results[(ticker, "MSE")] = Compare.MSE()
    all_results[(ticker, "QLIKE")] = Compare.QLIKE()


main.plot_table_all(all_results)
```
